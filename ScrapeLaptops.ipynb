```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# HTTP Request

Store website in variable

```python
website = 'https://www.laptopsdirect.co.uk/ct/laptops-and-netbooks/laptops'
```

Get request

```python
response = requests.get(website)
```

Status code

```python
response.status_code   # 200 - ok
```

# Soup object

```python
soup = BeautifulSoup(response.content, 'html.parser')     # create soup object to get access to all html elements
```

```python
soup
```

# Results

```python
results = soup.find_all('div', {'class':'OfferBox'})
```

## Results in Single Page

Put everything together inside a For-Loop

```python
product_name = []
product_price = []
product_relative_url = []
product_details = []

for result in results:
   #Name
    try:
        product_name.append(result.find('a', {'class': 'offerboxtitle'}).get_text())
    except:
        product_name.append('n/a')

   #price
    try:
        product_price.append(result.find('span', {'class': 'offerprice'}).get_text().strip())
    except:
        product_price.append('n/a')

    #relative URL
    try:
        product_relative_url.append(result.find('a', {'class':'offerboxtitle'}).get('href'))
    except:
        product_relative_url.append('n/a')

    #Product Details
    try:
        product_details.append(result.find('a', {'class': 'offerboxtitle'}).get_text())
    except:
        product_details.append('n/a')
```

```python
product_name
```

Combine URLs

```python
import urllib.parse

url_combined = []

for link in product_relative_url:
    url_combined.append(urllib.parse.urljoin(root_url, link))
```

```python
url_combined
```

Create Pandas Dataframe

```python
product_overview = pd.DataFrame({'Name': product_name, 'Price': product_price, 'Link': url_combined, 'Details': product_details})
product_overview
```

Output in Excel

```python
product_overview.to_excel('results_single_page.xlsx', index=False)
```
